```python
import os
import numpy as np
import matplotlib.pyplot as plt
CURRENT_DIRECTORY = os.getcwd()
DATA_AND_LAYER_NAME = "B_demag_125mT_0K_layer2"
DEMAG_B_Z_FILEPATH = os.path.join(CURRENT_DIRECTORY, "mumax_fields", "%s_z.npy" % DATA_AND_LAYER_NAME)
DEMAG_B_Z = np.load(DEMAG_B_Z_FILEPATH)
APPLIED_B_Z = 0.125
```

```python
plt.pcolormesh(DEMAG_B_Z, cmap="viridis")
plt.colorbar()
plt.title("B_z (T) under +125 mT external field")
```

```python
B_Z = DEMAG_B_Z + APPLIED_B_Z
plt.pcolormesh(B_Z, cmap="viridis")
plt.colorbar()
print(B_Z)
```
